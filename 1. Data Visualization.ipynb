```python
iris.quantile(0.1)#10th percentile
#The 𝑝th percentile is the number in the dataset such that roughly
#𝑝% of the data is less than this number. This number is also referred to as a quantile.
```

```python
iris.quantile(0.50) # 3rd quartile
```

```python
#If 𝑄𝑖 denotes the 𝑖th quartile, the inner-quartile range (IQR) is the difference between 
# the third quartile and the first quartile.
iris.quantile(0.75) - iris.quantile(.25)
```

```python
# SepalLengthCm	SepalWidthCm	PetalLengthCm	PetalWidthCm
# violin plot
#  The violinplot shows density of the length and width in the species.
# The thinner part denotes that there is less density whereas the fatter part conveys higher density
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sb.violinplot(x='Species',y='SepalLengthCm',data=iris)
plt.subplot(2,2,2)
sb.violinplot(x='Species',y='SepalWidthCm',data=iris)
plt.subplot(2,2,3)
sb.violinplot(x='Species',y='PetalLengthCm',data=iris)
plt.subplot(2,2,4)
sb.violinplot(x='Species',y='PetalWidthCm',data=iris)
```

```python
# Box PLot
# statistical five number summary 
# Minimum
# First Quartile
# Median (Second Quartile)
# Third Quartile
# Maximum
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sb.boxplot(x='Species',y='SepalLengthCm',data=iris)
plt.subplot(2,2,2)
sb.boxplot(x='Species',y='SepalWidthCm',data=iris)
plt.subplot(2,2,3)
sb.boxplot(x='Species',y='PetalLengthCm',data=iris)
plt.subplot(2,2,4)
sb.boxplot(x='Species',y='PetalWidthCm',data=iris)
```

```python
# One way we can extend this plot is adding a layer of individual points on top of
# it through Seaborn's striplot
# 
# We'll use jitter=True so that all the points don't fall in single vertical lines
# above the species
#
# Saving the resulting axes as ax each time causes the resulting plot to be shown
# on top of the previous axes
sb.boxplot(x="Species", y="PetalLengthCm", data=iris)
sb.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="red")
```

```python
# Correlation is a statistical measure that expresses the extent to which two variables are linearly related 
# (meaning they change together at a constant rate). It’s a common tool for describing simple relationships without 
# making a statement about cause and effect.

# we measure the correlation coefficient -1 to +1

# # Features of R
# The closer r is to zero, the weaker the linear relationship.
# Positive r values indicate a positive correlation, where the values of both variables tend to increase together.
# Negative r values indicate a negative correlation, where the values of one variable tend to increase when the values 
# of the other variable decrease.

iris.corr()
```
